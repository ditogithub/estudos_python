# Automação de Sistemas e Processos com Python

### Desafio:

Para controle de custos, todos os dias, seu chefe pede um relatório com todas as compras de mercadorias da empresa.
O seu trabalho, como analista, é enviar um e-mail para ele, assim que começar a trabalhar, com o total gasto, a quantidade de produtos compradas e o preço médio dos produtos.

E-mail do seu chefe: para o nosso exercício, coloque um e-mail seu como sendo o e-mail do seu chefe<br>
Link de acesso ao sistema da empresa: https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
#!pip install pyperclip
```

```python
import pyautogui
import pyperclip
import time

# Passo a Passo

pyautogui.PAUSE = 1
#----------------------------------------------------------------------------------------
# Passo 1 : Entrar no sistema da empresa (entrar no link de acesso)

pyautogui.hotkey("ctrl" , "t") # Abrir uma nova guia no navegador!
time.sleep(5)
pyautogui.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema") # Escrever o endereço do e-mail!
time.sleep(5)
pyautogui.press("enter") # Apértar enter para entrar na pagina do e-mail!
time.sleep(2)
pyautogui.click(x=591, y=376) # clicar na barra de login.
time.sleep(4)
#----------------------------------------------------------------------------------------
# Passo 2 : digitar login e senha de acesso

pyautogui.write("meulogin")
time.sleep(3)
pyautogui.click(x=588, y=453)
time.sleep(5)
pyautogui.write("minhasenha")
time.sleep(3)
pyautogui.click(x=673, y=524) # clicar em acessar!
time.sleep(3)
pyautogui.click(x=556, y=671 , clicks=2)  # Fechar o aviso de noteficações!
time.sleep(3)
#----------------------------------------------------------------------------------------
# Passo 3 : Exportar a base de dados.
pyautogui.click(x=486, y=407)  # Clicar no arquivo Compras.csv
time.sleep(3)
pyautogui.click(x=494, y=363) # Clicar nos Três pontinhos da tela.
time.sleep(3)
pyautogui.click(x=575, y=532) # Clicar em "Fazer Dowload do arquivo"
time.sleep(3)
pyautogui.press("enter") # Apertar "enter" em "Salvar" o Download
time.sleep(4)
pyautogui.hotkey("ctrl","w")
```

```python
# Passo 4 : Importar e Calcular os indicadores.

import pandas as pd  

tabela = pd.read_csv(r"C:\Users\Ricardo\Downloads\Compras.csv", sep=";")
display(tabela)
  
        # OBS: O comando "as" troca o nome "pandas" por "pd" para ficar mais fácil e rápido de programar e digitar.

        # OBS:A letra "r" no começo do endereço do arquivo csv, seguinifica que o python terá que reconhecer o o endereço do jeito que está escrito.
    
        # OBS: O Comando (sep = ";"), que seguinifica: separador de colunas, por padrão um arquivo csv vem com as colunas separadas por ","
          # mas esse veio separado com ";" , então esse comando informa ao python está no formato com ";" e organiza as colunas.

# ----------------------------------------------------------------------------------------------------------------------------------------------------

# Passo 5 : Calcular os indicadores.
total_gasto = tabela["ValorFinal"].sum()
quantidade = tabela["Quantidade"].sum()
preco_medio = total_gasto/quantidade
print("Total Gasto :", total_gasto)
print("Quantidade:",quantidade)  
print("Preço Médio:",preco_medio)
```

```python
import pyautogui
import pyperclip
import time

#Passo 6 : Enviar um e-mail para meu chefe.

#passo 1: Entrar no e-mail: https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox
    
pyautogui.hotkey("ctrl","t")  # Abre uma nova aba.
time.sleep(3)
pyperclip.copy("https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox") # Copia o link
time.sleep(3)
pyautogui.hotkey("ctrl","v") # Cola o link na barra de endereço do navegador.
time.sleep(3)
pyautogui.press("enter") # Aperta "Enter" para entrar na página.
time.sleep(3)

# passo 2: Clicar em escrever.
pyautogui.click(x=28, y=199)
time.sleep(7)
# passo 3: preencher as informações do e-mail.
# 1º Destinatário
pyautogui.write("ditopj@gmail.com") # Escreve o endereço de e-mail na barra de destinatário.
time.sleep(2)
pyautogui.press("tab") # seleciona e fixa o endereço na barra de destinatário.
time.sleep(2)
pyautogui.write("ricardao_machadopj@hotmail.com") # Escreve mais um endereço de e-mail na barra de destinatário.
time.sleep(2)
pyautogui.press("tab")
time.sleep(2)

# 2° Descrever o titulo assunto.
pyautogui.press("tab") # Aperta um tab para passar para a barra de endreço de assunto.
pyperclip.copy("Relatório de Vendas") # Digitando o titulo do assunto.
pyautogui.hotkey("ctrl","v")
pyautogui.press("tab") # Passando para o corpo do e-mail.

texto = f"""  
Prezados, 
segue o relatório de compras

Total Gasto:RS {total_gasto:,.2f}
Quantidade de produtos:{quantidade:,} 
Preço médio:RS {preco_medio:,.2f}

Qualquer dúvida, é so falar.
Att., Ricardo """

pyperclip.copy(texto)

pyautogui.hotkey("ctrl","v")
time.sleep(1)

# passo 4: Enviar   
pyautogui.hotkey("ctrl","enter")

# Fechar página e-mail

pyautogui.hotkey("ctrl","w")
```
